```
import pandas as pd

df=pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv')

df.columns=['country','beer','spirit','wine','total_litres']

asean_country= [
                'Thailand',
                'Vietnam',
                'Malaysia',
                'Brunei',
                'Laos',
                'Myanmar',
                'Singapore',
                'Cambodia',
                'Indonesia',
                'Philippines',
                ]
eu_country = [
              'Austria',	'Belgium', 'Bulgaria', 'Croatia', 'Cyprus',
              'Czechia',	'Denmark', 'Estonia', 'Finland', 'France',
              'Germany',	'Greece', 'Hungary', 'Ireland','Italy',	
              'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands',
              'Poland','Portugal','Romania','Slovakia','Slovenia','Spain','Sweden',]


asean_df = df[ df['country'].isin(asean_country) ]
print(asean_df)
pd.pivot_table(asean_df.drop(labels='total_litres',axis=1), index='country').plot(kind='bar',figsize=(15,15))
```

```
import plotly.express as px
pd.set_option("display.max_rows", None, "display.max_columns", None)
location_data = px.data.gapminder().query("year==2007")
inject_data = pd.DataFrame({'country':['Laos','South Korea'], 
                            'continent':['Asia','Asia'], 
                            'year':['2007','2007'],	
                            'lifeExp':[72.390,72.390],	
                            'pop': [190010647,190010647], 
                            'gdpPercap':[9065.800825,9065.800825],	
                            'iso_alpha':['LAO','KOR'],	
                            'iso_num': ['76','76']
                          })
location_data = location_data.append( inject_data )
location_data.at[1619,'country']='USA'
located_df = pd.merge(df, location_data, how='left', on='country')

located_df



```

```
located_df=located_df.dropna()
fig = px.scatter_geo( located_df, locations="iso_alpha", color="continent",
                     hover_name="country", size="spirit",
                     projection="natural earth", opacity=1.0)
fig.show()
```

```
fig = px.scatter_geo( located_df, locations="iso_alpha", color="continent",
                     hover_name="country", size="beer",
                     projection="natural earth", opacity=1.0)
fig.show()
```
